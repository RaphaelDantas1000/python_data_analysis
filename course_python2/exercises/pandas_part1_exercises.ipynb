9) For men, a BMI of 20 to 24.9 is considered normal. Above that, it is considered overweight, and below it is considered underweight. Create a new column called "bmi_male" that will receive the text "below normal" for BMI below 20, "normal" for BMI from 20 to 24.9, and "above normal" for BMI above 24.9.

```python
conditions = [
    (df["sex"] == "male") & (df["bmi"] < 20),
    (df["sex"] == "male") & (df["bmi"] >= 20) & (df["bmi"] <= 24.9),
    (df["sex"] == "male") & (df["bmi"] > 24.9)
]

choices = ["below normal", "normal", "above normal"]

df["bmi_male"] = np.select(conditions, choices, default="not male")
```
